## Linked List

```python
## Define a linked list class in python
class LinkedList:
    def __init__(self, val = 0, next = None):
        self.val = val
        self.next = next
```

```python
## Produce an instance
node3 = LinkedList(3)
node2 = LinkedList(2, node3)
node1 = LinkedList(1, node2)
```

```python
# This gives the value
print(node1.val)

# This gives the memory address of the next LinkedList after node1 (which would be node2)
print(node1.next)
```

So why do we care about Linked Lists at all? Surely it does the same thing as arrays?

### Insertion/Deletion
Compared to arrays, linked lists has $O(1)$ insertion/deletion at front, compared to $O(n)$ for arrays.

```python
# Consider array insertion
x = [1,2,3]

# Insert 0 at front would be O(n) as it would have to move the pointers/references. 
x.insert(0, 0)
print(x)

# For linked lists, this would be O(1)
head = LinkedList(0, next=node1)
print(head.val)
print(head.next.val)
```
